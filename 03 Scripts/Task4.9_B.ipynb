### 3.You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column. In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
df.dtypes
```

```python
df['order_hour_of_day'] = df['order_hour_of_day'].astype('float64')
```

```python
hist=df['order_hour_of_day'].plot.hist(bins = 25)
plt.xlabel('Hour of the day')
plt.ylabel('Sales')
plt.title('Number of sales per hour')
```

### 4.The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
bar=df['loyalty_flag'].value_counts().plot.bar()
plt.xlabel('Loyalty')
plt.ylabel('Relative frequencies of Sales')
plt.title('Frequencies of sales per type of loyalty')
```
